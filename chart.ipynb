# Red Squirrel Diet Study Results
We gave 100 Red Squirrels one of three diets and recorded their adult weights. We'll look at how each diet may have effected their wieghts.

```python
import pandas as pd
from plotly import graph_objects as go
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read Data
    # Note: I can't access the Jupyter Notebook on the Practicum platform and the Project Instructions 
    # do not mention the path to the squirrel_diet.csv file, so I have no way of figuring that out. 
    # I can easily update my code's structure once I have that information, if needed.
sq_df = pd.read_csv('squirrel_diet.csv')

# Change Numeric Data into String Data, for readability later
sq_df['sex_str'] = sq_df['sex'].replace({0: 'Female', 1: 'Male'})
sq_df['diet_str'] = sq_df['diet'].replace({1: 'A', 2: 'B', 3: 'C'})

# sq_df.info()
```

```python
# print(sq_df.sample(10))
```

```python
# Set up the style
sns.set(
    style='white',
    font='Comic Sans MS',
    palette=['magenta', 'green']
)

# Make plot
stripplot = sns.stripplot(
    data=sq_df, 
    x='diet_str', 
    y='weight', 
    hue='sex_str'
)

# Set title, x-label, and y-label
plt.title('Red Squirrel Diet Study')
plt.xlabel('Diet Label')
plt.ylabel('Adult Weight / lbs')

# Reverse the x-axis
ax = plt.gca()
ax.invert_xaxis()

# Show tick marks on the axes
ax.tick_params(bottom=True, left=True)

# Set legend title and show legend
legend = stripplot.legend(title='sex')
legend.set_visible(True)

plt.show()
```

It looks like the B Diet is a little all over the place, but the A Diet and, to a slightly lesser degree, the C Diet resulted in lighter squirrels.
